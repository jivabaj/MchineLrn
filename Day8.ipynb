```python
#classification using decision Tree

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('/content/Social_Network_Ads.csv')

x = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test, y_train,y_test = train_test_split(x, y, test_size = 0.20)

print(x_train)
print(x_test)
print(y_train)
print(y_test)
```

```python
#x_train will be fit and tranfroemed, x_test will only be transformed

from sklearn.preprocessing import StandardScaler

sc = StandardScaler()

x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)

print(x_test)
print(x_train)
```

```python
#Import decision tree classifier and train decision tree for our training data

from sklearn.tree import DecisionTreeClassifier

classifier = DecisionTreeClassifier(criterion  = 'entropy')  #method to implement gini method

classifier.fit(x_train, y_train)
```

```python
#Testing the classifier with test set data

y_pred = classifier.predict(x_test)
```

```python
#Test and preduct with a new result

print(classifier.predict(sc.transform([[30,87000]])))
print(classifier.predict(sc.transform([[30,187000]])))
```

```python
#Build confusion matrix and get accuracy of our model for 25% test

from sklearn.metrics import confusion_matrix, accuracy_score

cm = confusion_matrix(y_pred, y_test)

print(cm)

accuracy_score(y_pred, y_test)
```

```python
#for 20% should be more than 25%

from sklearn.metrics import confusion_matrix, accuracy_score

cm = confusion_matrix(y_pred, y_test)

print(cm)

accuracy_score(y_pred, y_test)
```

```python
#Visualizing training set results

from matplotlib.colors import ListedColormap
x_set, y_set = sc.inverse_transform(x_train), y_train  # transforming the neg values of x as outliers are present. No need for y_train tranforming

x1, x2 = np.meshgrid(np.arange(start = x_set[:, 0].min() - 0, stop = x_set[:, 0].max() + 0, step = 0.25), 
                     np.arange(start = x_set[:, 1].min() - 0, stop = x_set[:, 1].max() + 1, step = 0.25)) # rep max and min variables

plt.contourf(x1, x2, classifier.predict(sc.transform(np.array([x1.ravel(), x2.ravel()]).T)).reshape(x1.shape), cmap = ListedColormap(('red', 'green')))

plt.xlim(x1.min(), x1.max())
plt.ylim(x2.min(), x2.max())

for i,j in enumerate(np.unique(y_set)):  # j represents 0 or 1
  plt.scatter(x_set[y_set == j, 0], x_set[y_set == j, 1], c = ListedColormap(('red', 'green'))(i), label = j)

plt.title('Decision TRee Classification (training set)')
plt.xlabel('Age')
plt.ylabel('Salary')
plt.legend()
plt.show()
```

```python
#using gini criterion now

from sklearn.tree import DecisionTreeClassifier

classifier = DecisionTreeClassifier(criterion  = 'gini', max_depth = 3, min_samples_leaf= 5)  

classifier.fit(x_train, y_train)
```

```python
#predicting test set results

y_pred = classifier.predict(x_test)
```

```python
print(classifier.predict(sc.transform([[30,87000]])))
print(classifier.predict(sc.transform([[30,187000]])))
```

```python
#confusion matrix and accuracy

cm = confusion_matrix(y_pred, y_test)
print(cm)

accuracy_score(y_pred, y_test)
```
