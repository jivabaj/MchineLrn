```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('/content/Social_Network_Ads.csv')

x = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values

print(x)
print(y)
```

```python
#Split into train and test

from sklearn.model_selection import train_test_split

x_train,x_test, y_train,y_test = train_test_split(x, y, test_size = 0.25)

print(x_train)
print(x_test)
print(y_test)
print(y_train)
```

```python
# Feature scaling, difference of range between the variables age and salary. so we have to do standard scaling for standard representation of all data. without range difference

from sklearn.preprocessing import StandardScaler

sc = StandardScaler()

x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)

print(x_train)
print(x_test)
```

```python
#Logistic Regression Implementation

from sklearn.linear_model import LogisticRegression

classifier = LogisticRegression()

classifier.fit(x_train, y_train)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
```

```python
# Predict the test set result

y_pred = classifier.predict(x_test)
```

```python
# Predict the new entry

print(classifier.predict(sc.transform([[20, 850000]])))
print(classifier.predict(sc.transform([[30, 85000]])))
```

```python
cm = confusion_matrix(y_pred, y_test)
print(cm)

accuracy_score(y_pred, y_test)
```

```python
#Visualizing training set results

from matplotlib.colors import ListedColormap
x_set, y_set = sc.inverse_transform(x_train), y_train  # transforming the neg values of x as outliers are present. No need for y_train tranforming

x1, x2 = np.meshgrid(np.arange(start = x_set[:, 0].min() - 0, stop = x_set[:, 0].max() + 0, step = 0.25), np.arange(start = x_set[:, 1].min() - 0, stop = x_set[:, 1].max() + 1, step = 0.25)) # rep max and min variables

plt.contourf(x1, x2, classifier.predict(sc.transform(np.array([x1.ravel(), x2.ravel()]).T)).reshape(x1.shape), cmap = ListedColormap(('red', 'green')))

plt.xlim(x1.min(), x1.max())
plt.ylim(x2.min(), x2.max())

for i,j in enumerate(np.unique(y_set)):  # j represents 0 or 1
  plt.scatter(x_set[y_set == j, 0], x_set[y_set == j, 1], c = ListedColormap(('red', 'green'))(i), label = j)

plt.title('Logistic Regression (training set)')
plt.xlabel('Age')
plt.ylabel('Salary')
plt.legend()
plt.show()
```
