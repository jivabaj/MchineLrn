# **Matplotlib**

```python
#iris dataset

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#working with default datasets

data = sns.load_dataset('iris')

data.head()
```

```python
#matplotlib

#line graph

plt.plot(data['sepal_width'])
plt.show()  #print the graph
```

```python
plt.plot(data['petal_length'])
plt.title("Line Chart using petal_length")
plt.xlabel('Indexes')
plt.ylabel('petal length')
plt.show()
```

```python
plt.plot(data['sepal_width']) 
plt.title('Iris Species', fontsize = 27)
plt.xlabel('Indexes')
plt.ylabel('sepal width')

plt.show()
```

```python
plt.plot(data['sepal_width'], color = 'black') 
plt.title('Iris Species', fontsize = 27)
plt.xlabel('Indexes')
plt.ylabel('sepal width')

plt.show()
```

```python
#sepal length, sepal width. petal l and w in the same graph

plt.plot(data['sepal_width'], color = 'red', label = 'sepal_width')  #label = legend
plt.plot(data['sepal_length'], color = 'green', label = 'sepal lenth')
plt.plot(data['petal_length'], color = 'black', label = 'petal length')
plt.plot(data['petal_width'], color = 'yellow', label = 'petal width')
plt.title('Iris Species', fontsize = 27)
plt.xlabel('Indexes')
plt.ylabel('Features')

plt.show()
```

```python
#fixed size

plt.figure(figsize = (8,4))
plt.plot(data['sepal_width'], color = 'red', label = 'sepal_width')  #label = legend
plt.plot(data['sepal_length'], color = 'green', label = 'sepal lenth')
plt.plot(data['petal_length'], color = 'black', label = 'petal length')
plt.plot(data['petal_width'], color = 'yellow', label = 'petal width')
plt.title('Iris Species', fontsize = 27)
plt.xlabel('Indexes')
plt.ylabel('Features')
plt.legend()  #to show the legend
plt.show()
```

```python
#Adding dashed lines

plt.plot(data['sepal_width'], color = 'red', label = 'sepal_width', linestyle = "--")  #label = legend
plt.plot(data['sepal_length'], color = 'green', label = 'sepal lenth', linestyle = "--")
plt.plot(data['petal_length'], color = 'black', label = 'petal length')
plt.plot(data['petal_width'], color = 'yellow', label = 'petal width')
plt.title('Iris Species', fontsize = 27)
plt.xlabel('Indexes')
plt.ylabel('Features')
plt.legend()
plt.show()
```

```python
plt.figure(figsize = (12,6))
plt.plot(data['sepal_width'], color = 'red', label = 'sepal_width', linestyle = "--")  #label = legend
plt.plot(data['sepal_length'], color = 'green', label = 'sepal lenth', linestyle = "--")
plt.plot(data['petal_length'], color = 'black', label = 'petal length', marker = '*')  # shows at the peaks
plt.plot(data['petal_width'], color = 'yellow', label = 'petal width', marker = '+')
plt.title('Iris Species', fontsize = 27)
plt.xlabel('Indexes')
plt.ylabel('Features')
plt.legend()
plt.show()
```

```python
plt.figure(figsize = (12,6))
plt.plot(data['sepal_width'], color = 'red', label = 'sepal_width', linestyle = "--")  #label = legend
plt.plot(data['sepal_length'], color = 'green', label = 'sepal lenth', linestyle = "--")
plt.plot(data['petal_length'], color = 'black', label = 'petal length', marker = '*')
plt.plot(data['petal_width'], color = 'yellow', label = 'petal width', marker = '+', linewidth = 3) # changing the width of line
plt.title('Iris Species', fontsize = 27)
plt.xlabel('Indexes')
plt.ylabel('Features')
plt.legend()
plt.show()
```

```python
plt.figure(figsize = (12,6))

plt.grid()  #adding grids to graph

plt.plot(data['sepal_width'], color = 'red', label = 'sepal_width', linestyle = "--")  #label = legend
plt.plot(data['sepal_length'], color = 'green', label = 'sepal lenth', linestyle = "--")
plt.plot(data['petal_length'], color = 'black', label = 'petal length', marker = '*')
plt.plot(data['petal_width'], color = 'yellow', label = 'petal width', marker = '+') 
plt.title('Iris Species', fontsize = 27)
plt.xlabel('Indexes')
plt.ylabel('Features')
plt.legend()
plt.show()
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
#Histogram

plt.hist(tips['total_bill'], bins = 20, edgecolor = 'black')
plt.show()
```

```python
#piechart

day = tips.groupby('day')['total_bill'].sum().reset_index()
x = day['day']
y = day['total_bill']
explode = [0,0,0,0.1]
plt.pie(y, labels = x, explode = explode, autopct = '%1.1f%%', colors = ['red','blue', 'green', 'yellow'], wedgeprops = {'edgecolor' : 'black'})  #shadow ?
plt.show()
```

```python
#box plot

plt.boxplot(tips['total_bill'])
plt.show()
```

```python
#scatter plot

plt.scatter(tips['total_bill'], tips['tip'])
plt.title('Scatter Plot')
plt.xlabel('Total Bill')
plt.ylabel('Tips')
plt.show()
```

```python
plt.scatter(tips['total_bill'], tips['tip'], c = 'red') # c=colour
plt.title('Scatter Plot')
plt.xlabel('Total Bill')
plt.ylabel('Tips')
plt.show()
```

```python
plt.scatter(tips['total_bill'], tips['tip'], c = tips['size'])  #color changes according to the size of tip
plt.title('Scatter Plot')
plt.xlabel('Total Bill')
plt.ylabel('Tips')

cbar = plt.colorbar()
cbar.set_label('Size')

plt.show()
```

```python
# Bar plot

sns.barplot(x = 'day', y = 'total_bill', data = tips)
```

```python
sns.distplot(tips['total_bill'], hist = 'false')
```
